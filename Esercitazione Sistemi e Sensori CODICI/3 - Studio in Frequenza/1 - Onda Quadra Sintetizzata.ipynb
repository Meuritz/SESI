
### sintetizzare un segnale di treno di impulsi (onda quadra) a partire dai suoi coefficienti di Fourier e plottarne la waveform

![image](img/image.png)

#### f0 = frenquenza fondamentale del onda quadra
#### t = vettore del tempo (tempo reale)

```python
import numpy as np
from scipy import signal, fft, fftpack
import matplotlib.pyplot as plt
```

## Array per il tempo 

```python
#tempo di campionamento
DURATA = 1

#secondi di campionamento
FS = 500

#array per il tempo
t = np.linspace(0, DURATA, int(DURATA * FS), endpoint=False)

# inizializzo k
k = 1
```

## Calcolo dei coefficienti

```python
risultato = []

# numero coefficienti da calcolare
nc = 1000

#calcolo dei coefficienti
while k <= nc:
    
    # calcola il coefficiente di Fourier
    coefficiente = (1 / k) * ((-1) ** ((k - 1) / 2)) * np.cos(2 * np.pi * k * t / DURATA)
    
    # aggiungo il coefficente appena calcolato al risultato 
    risultato.append((2 / np.pi) * coefficiente)

    # aumento k per la prossima iterazione
    k += 2
```

## Plot dell'onda

```python
# Sommo gli array ottenuti 
waveform = np.sum(risultato, axis=0)

# Plot del onda
plt.plot(t, waveform, "purple")
plt.xlabel('Tempo')
plt.ylabel('Ampiezza')
plt.title('Onda Quadra Sintetizzata')
plt.text(0, 0, f"N Coefficienti:\n         {str(nc)}", fontsize="10")
plt.show()
```
