## Studio onde a diversi hz 

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import fft, signal
```

## Varibili per creare un segnale nel tempo

```python
# Frenquenza in Hz
freq = 0
#Durata in secondi
durata = 1
#frequenza campionamento (al secondo)
fs = 100
#array per il tempo
t = np.linspace(0, durata, int(fs*durata))
```

## Onda Sinusoidale

### 100 hz

```python
freq = 100
plt.plot(t ,np.sin(2*np.pi*freq*t))
plt.show()
```

### 200 hz

```python
freq = 200
plt.plot(t ,np.sin(2*np.pi*freq*t))
plt.show()
```

### 440 hz

```python
freq = 440
plt.plot(t ,np.sin(2*np.pi*freq*t))
plt.show()
```

## Onda Triangolare

### 100 hz

```python
freq = 100
plt.plot(t, signal.sawtooth(2*np.pi*freq*t, 0.5))
plt.show()
```

### 200 hz
