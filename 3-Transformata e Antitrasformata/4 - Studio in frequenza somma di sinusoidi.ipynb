## Studio in frequenza di un onda fatta dalla somma di 3 Sinusoidi con frequenza diversa

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import fft, signal
```

### Varibili per creare le onde nel tempo

```python
# Frenquenza in Hz
freq = 0
#Durata in secondi
durata = 1
#frequenza campionamento (al secondo)
fs = 100
#array per il tempo
t = np.linspace(0, durata, int(fs*durata))
```

## Creazione delle onde e plot

```python
#creo le onde a vari hz
freq = 100
sin1 = np.sin(2*np.pi*freq*t)

freq = 200
sin2 = np.sin(2*np.pi*freq*t)

freq = 440
sin3 = np.sin(2*np.pi*freq*t)

# Somma delle 3 onde
sin_sum = sin1 + sin2 + sin3

# Trasformata onda Sin 
ft_sin_sum = fft.fft(sin_sum)

# plot
fig, ax = plt.subplots(ncols=3, figsize=(20, 5))

# Onda nel tempo
ax[0].plot(t, sin_sum, "r")
ax[0].set_title(f"Onda Sinusoidale 100,200,440")
ax[0].set_xlabel("Tempo (s)")
ax[0].set_ylabel("Ampiezza")

# Parte reale ft (pari)
ax[1].plot(np.fft.fftshift(np.real(ft_sin_sum)), "b")
ax[1].set_title("Parte Reale")
ax[1].set_xlabel("Indice Frequenza")
ax[1].set_ylabel("Ampiezza")

# Parte immaginaria ft (dispari)
ax[2].plot(np.fft.fftshift(np.imag(ft_sin_sum)), "g")
ax[2].set_title("Parte Immaginaria")
ax[2].set_xlabel("Indice Frequenza")
ax[2].set_ylabel("Ampiezza")

plt.show()
```
